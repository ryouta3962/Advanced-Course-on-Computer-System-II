```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
import japanize_matplotlib  # 日本語表示ライブラリ

# --- 1. データの読み込みと前処理 ---
try:
    df = pd.read_csv('sensor_features.csv')
except FileNotFoundError:
    print("エラー: 'sensor_features.csv' が見つかりません。")
    print("先に 'calculate_features.py' を実行してください。")
    exit()

print("データの読み込みが完了しました。")

features_df = df.drop(columns=['filename', 'label'])
true_labels = df['label']

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)
print("特徴量の標準化が完了しました。")

# --- 2. K-meansクラスタリングの実行 ---
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
predicted_clusters = kmeans.fit_predict(features_scaled)
df['cluster'] = predicted_clusters
print(f"K-meansクラスタリング (k={n_clusters}) が完了しました。")


# --- 3. 結果の評価 (クロス集計) ---
crosstab = pd.crosstab(df['label'], df['cluster'])
print("\n--- クラスタリング結果のクロス集計 ---")
print(crosstab)
print("\n[見方]")
print("・各行が実際の行動ラベル、各列がK-meansによって割り当てられたクラスタ番号です。")
print("・例えば、'aruki'行のデータが'cluster 0'列に集中していれば、'cluster 0'は歩行動作を表していると解釈できます。")


# --- 4. 正答率の計算 ---
print("\n--- 正答率の計算 ---")
# 1. クラスタ番号と実際のラベルを対応付けるマップを作成
cluster_to_label_map = {}
# crosstabの各列（クラスタ）に対してループ
for cluster_num in crosstab.columns:
    # そのクラスタで最も頻繁に出現するラベルを特定
    most_frequent_label = crosstab[cluster_num].idxmax()
    cluster_to_label_map[cluster_num] = most_frequent_label

print("クラスタとラベルの対応付け:")
print(cluster_to_label_map)

# 2. 予測されたクラスタ番号を、対応付けたラベル名に変換
predicted_labels = df['cluster'].map(cluster_to_label_map)

# 3. 実際のラベルと予測ラベルを比較して正答率を計算
accuracy = accuracy_score(true_labels, predicted_labels)
print(f"\n正答率: {accuracy:.2%}") # パーセント表示


# --- 5. 結果の可視化 (PCAによる次元削減) ---
print("\n結果を可視化しています...")
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)

df_pca = pd.DataFrame(data=features_pca, columns=['PC1', 'PC2'])
df_pca['cluster'] = df['cluster']
df_pca['label'] = df['label']

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

sns.scatterplot(ax=axes[0], x='PC1', y='PC2', hue='cluster', data=df_pca, palette='viridis', s=100, alpha=0.9)
axes[0].set_title('K-meansによるクラスタリング結果 (PCAで2次元に可視化)', fontsize=14)
axes[0].set_xlabel('主成分1', fontsize=12)
axes[0].set_ylabel('主成分2', fontsize=12)
axes[0].legend(title='予測クラスタ')
axes[0].grid(True)

sns.scatterplot(ax=axes[1], x='PC1', y='PC2', hue='label', data=df_pca, palette='deep', s=100, alpha=0.9)
axes[1].set_title('実際の行動ラベル (PCAで2次元に可視化)', fontsize=14)
axes[1].set_xlabel('主成分1', fontsize=12)
axes[1].set_ylabel('主成分2', fontsize=12)
axes[1].legend(title='実際の行動')
axes[1].grid(True)

plt.suptitle(f'K-meansクラスタリングの結果と実際のラベルの比較 (正答率: {accuracy:.2%})', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

print("\n分析が完了しました。")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier  # RandomForestを追加
import japanize_matplotlib  # 日本語表示ライブラリ

# --- 1. データの読み込みと前処理 ---
try:
    df = pd.read_csv('sensor_features.csv')
except FileNotFoundError:
    print("エラー: 'sensor_features.csv' が見つかりません。")
    print("先に 'calculate_features.py' を実行してください。")
    exit()

print("データの読み込みが完了しました。")

features_df = df.drop(columns=['filename', 'label'])
true_labels = df['label']

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)
print("特徴量の標準化が完了しました。")


# ===================================================================
# ここからK-meansの分析（変更なし）
# ===================================================================

# --- 2. K-meansクラスタリングの実行 ---
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
predicted_clusters = kmeans.fit_predict(features_scaled)
df['cluster'] = predicted_clusters
print(f"\nK-meansクラスタリング (k={n_clusters}) が完了しました。")


# --- 3. 結果の評価 (クロス集計) ---
crosstab = pd.crosstab(df['label'], df['cluster'])
print("\n--- クラスタリング結果のクロス集計 ---")
print(crosstab)


# --- 4. 正答率の計算 ---
print("\n--- 正答率の計算 ---")
cluster_to_label_map = {}
for cluster_num in crosstab.columns:
    most_frequent_label = crosstab[cluster_num].idxmax()
    cluster_to_label_map[cluster_num] = most_frequent_label
print("クラスタとラベルの対応付け:", cluster_to_label_map)
predicted_labels = df['cluster'].map(cluster_to_label_map)
accuracy = accuracy_score(true_labels, predicted_labels)
print(f"\n正答率: {accuracy:.2%}")


# --- 5. 結果の可視化 (PCAによる次元削減) ---
print("\nK-meansの結果を可視化しています...")
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)
df_pca = pd.DataFrame(data=features_pca, columns=['PC1', 'PC2'])
df_pca['cluster'] = df['cluster']
df_pca['label'] = df['label']
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.scatterplot(ax=axes[0], x='PC1', y='PC2', hue='cluster', data=df_pca, palette='viridis', s=100, alpha=0.9)
axes[0].set_title('K-meansによるクラスタリング結果 (PCAで2次元に可視化)', fontsize=14)
# ... (以下、グラフ表示のコードは省略)
sns.scatterplot(ax=axes[1], x='PC1', y='PC2', hue='label', data=df_pca, palette='deep', s=100, alpha=0.9)
axes[1].set_title('実際の行動ラベル (PCAで2次元に可視化)', fontsize=14)
plt.suptitle(f'K-meansクラスタリングの結果と実際のラベルの比較 (正答率: {accuracy:.2%})', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


# ===================================================================
# ここからランダムフォレストによる特徴量重要度の分析（新規追加）
# ===================================================================

print("\n--- ランダムフォレストによる特徴量の重要度分析 ---")

# 1. ランダムフォレストモデルの作成と訓練
#    教師あり学習なので、標準化前のデータと正解ラベルを使用
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(features_df, true_labels)

# 2. 特徴量の重要度を取得
importances = rf_model.feature_importances_
feature_names = features_df.columns

# 重要度をDataFrameにまとめて、降順でソート
feature_importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values('importance', ascending=False)

print("特徴量の重要度 (上位10件):")
print(feature_importance_df.head(10))

# 3. 重要度の可視化
print("\n特徴量の重要度を可視化しています...")
plt.figure(figsize=(12, 10))
sns.barplot(x='importance', y='feature', data=feature_importance_df, palette='rocket')
plt.title('ランダムフォレストによる特徴量の重要度', fontsize=16)
plt.xlabel('重要度 (Importance)', fontsize=12)
plt.ylabel('特徴量 (Feature)', fontsize=12)
plt.tight_layout()
plt.show()

print("\nすべての分析が完了しました。")
```
