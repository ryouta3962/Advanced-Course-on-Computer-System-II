```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
import japanize_matplotlib  # 日本語表示ライブラリ

# --- 1. データの読み込みと前処理 ---
# ★変更点: 読み込むファイルをタイムウィンドウ版に変更
INPUT_FILE = 'sensor_features_windowed.csv'
try:
    df = pd.read_csv(INPUT_FILE)
except FileNotFoundError:
    print(f"エラー: '{INPUT_FILE}' が見つかりません。")
    print("先に 'calculate_features_windowed.py' を実行してください。")
    exit()

print(f"データの読み込みが完了しました。データ数: {len(df)}件")

# 特徴量として使用するデータを抽出
features_df = df.drop(columns=['filename', 'label'])
true_labels = df['label']

# データの標準化
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)
print("特徴量の標準化が完了しました。")


# --- 2. K-meansクラスタリングの実行 ---
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
predicted_clusters = kmeans.fit_predict(features_scaled)
df['cluster'] = predicted_clusters
print(f"K-meansクラスタリング (k={n_clusters}) が完了しました。")


# --- 3. 結果の評価 (クロス集計) ---
crosstab = pd.crosstab(df['label'], df['cluster'])
print("\n--- クラスタリング結果のクロス集計 ---")
print(crosstab)
print("\n[見方]")
print("・各行が実際の行動ラベル、各列がK-meansによって割り当てられたクラスタ番号です。")


# --- 4. 正答率の計算 ---
print("\n--- 正答率の計算 ---")
cluster_to_label_map = {}
for cluster_num in crosstab.columns:
    most_frequent_label = crosstab[cluster_num].idxmax()
    cluster_to_label_map[cluster_num] = most_frequent_label

print("クラスタとラベルの対応付け:")
print(cluster_to_label_map)

predicted_labels = df['cluster'].map(cluster_to_label_map)
accuracy = accuracy_score(true_labels, predicted_labels)
print(f"\n正答率: {accuracy:.2%}")


# --- 5. 結果の可視化 (PCAによる次元削減) ---
print("\n結果を可視化しています...")
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)

df_pca = pd.DataFrame(data=features_pca, columns=['PC1', 'PC2'])
df_pca['cluster'] = df['cluster']
df_pca['label'] = df['label']

# データ点が多いので、プロットの透明度とサイズを調整
plot_alpha = 0.5
plot_size = 30

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# プロット1: K-meansによるクラスタリング結果
sns.scatterplot(ax=axes[0], x='PC1', y='PC2', hue='cluster', data=df_pca, 
                palette='viridis', s=plot_size, alpha=plot_alpha)
axes[0].set_title('K-meansクラスタリング結果 (タイムウィンドウデータ)', fontsize=14)
axes[0].set_xlabel('主成分1', fontsize=12)
axes[0].set_ylabel('主成分2', fontsize=12)
axes[0].legend(title='予測クラスタ')
axes[0].grid(True)

# プロット2: 実際のラベル（正解データ）
sns.scatterplot(ax=axes[1], x='PC1', y='PC2', hue='label', data=df_pca, 
                palette='deep', s=plot_size, alpha=plot_alpha)
axes[1].set_title('実際の行動ラベル (タイムウィンドウデータ)', fontsize=14)
axes[1].set_xlabel('主成分1', fontsize=12)
axes[1].set_ylabel('主成分2', fontsize=12)
axes[1].legend(title='実際の行動')
axes[1].grid(True)

plt.suptitle(f'タイムウィンドウデータに対するK-means結果 (正答率: {accuracy:.2%})', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

print("\n分析が完了しました。")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix, adjusted_rand_score
# from scipy.stats import mode # ★ 不要になったので削除またはコメントアウト

# --- 設定項目 ---
INPUT_FILE = 'sensor_features_window.csv'
N_CLUSTERS = 3  # クラスタ数 (aruki, kudari, nobori の3種類)

# --- 1. データ読み込みと準備 ---
print(f"--- 1. データの読み込みと準備 ---")
try:
    df = pd.read_csv(INPUT_FILE)
except FileNotFoundError:
    print(f"エラー: 入力ファイル '{INPUT_FILE}' が見つかりません。")
    print("先に特徴量抽出のスクリプトを実行してください。")
    exit()

# 特徴量 (X) と 正解ラベル (y_true) に分割
# 'source_file' は分析に使用しないため除外
X = df.drop(['label', 'source_file'], axis=1)
y_true = df['label']

# ラベルのユニークな値を取得（後のプロットや評価で使用）
labels_true_unique = sorted(y_true.unique())

print(f"データセットの形状: {df.shape}")
print(f"特徴量の数: {X.shape[1]}")
print(f"正解ラベルの種類: {labels_true_unique}")
print("-" * 30, "\n")


# --- 2. 前処理 (標準化) ---
print(f"--- 2. 前処理 (標準化) ---")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("特徴量の標準化が完了しました。")
print("-" * 30, "\n")


# --- 3. K-meansクラスタリングの実行 ---
print(f"--- 3. K-meansクラスタリングの実行 (k={N_CLUSTERS}) ---")
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=42, n_init=10)
y_pred_cluster = kmeans.fit_predict(X_scaled) # 予測されたクラスタラベル (0, 1, 2)
df['cluster'] = y_pred_cluster
print("K-meansによるクラスタリングが完了しました。")
print("-" * 30, "\n")


# --- 4. 結果の評価 ---
print(f"--- 4. 結果の評価 ---")

# 4-1. K-meansクラスタと正解ラベルの対応付け
# ★★★ エラー箇所を修正 ★★★
labels_pred_mapped = np.zeros_like(y_pred_cluster, dtype=object)
mapping = {}
for i in range(N_CLUSTERS):
    mask = (y_pred_cluster == i)
    # クラスタにデータが割り当てられているかチェック
    if np.any(mask):
        # 該当クラスタ内の正解ラベルで最も頻度の高いものを取得
        most_common_label = y_true[mask].value_counts().idxmax()
        labels_pred_mapped[mask] = most_common_label
        mapping[i] = most_common_label
    else:
        # 稀にクラスタが空になるケースへの対応
        mapping[i] = "Unassigned" 

print("クラスタと正解ラベルの対応付け:")
print(mapping)

# 4-2. 正答率 (Accuracy) の計算
accuracy = accuracy_score(y_true, labels_pred_mapped)
print(f"\n正答率 (Accuracy): {accuracy:.4f} ({accuracy*100:.2f}%)")

# 4-3. Adjusted Rand Index (ARI) の計算 (ラベル対応付け不要の評価指標)
ari = adjusted_rand_score(y_true, y_pred_cluster)
print(f"Adjusted Rand Index (ARI): {ari:.4f}")

# 4-4. 混同行列 (Confusion Matrix) の表示
# ラベルがマッピングされなかったケースも考慮
conf_mat_labels = sorted(list(mapping.values()))
conf_mat = confusion_matrix(y_true, labels_pred_mapped, labels=labels_true_unique)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', 
            xticklabels=labels_true_unique, yticklabels=labels_true_unique)
plt.title('Confusion Matrix (K-means)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('confusion_matrix_kmeans.png') # ファイルに保存
print("\n混同行列を 'confusion_matrix_kmeans.png' として保存しました。")
print("-" * 30, "\n")


# --- 5. 可視化 (PCAによる次元削減) ---
print(f"--- 5. 結果の可視化 ---")
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 可視化用のDataFrameを作成
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['True Label'] = y_true
df_pca['K-means Cluster'] = df['cluster'].map(mapping) # 対応付け後のラベル名を使用

plt.figure(figsize=(16, 7))

#  subplot 1: 元の正解ラベルによるプロット
plt.subplot(1, 2, 1)
sns.scatterplot(x='PC1', y='PC2', hue='True Label', hue_order=labels_true_unique, data=df_pca, palette='viridis', alpha=0.8, s=50)
plt.title('Original True Labels (after PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)

# subplot 2: K-meansの結果によるプロット
plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='K-means Cluster', hue_order=labels_true_unique, data=df_pca, palette='viridis', alpha=0.8, s=50)
plt.title('K-means Clustering Results (after PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)

plt.suptitle('K-means Clustering vs True Labels', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('kmeans_vs_true_labels_pca.png') # ファイルに保存
print("PCAによる可視化結果を 'kmeans_vs_true_labels_pca.png' として保存しました。")

plt.show() # プロットを表示
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix, adjusted_rand_score

# --- 設定項目 ---
INPUT_FILE = 'sensor_features_window.csv'
N_CLUSTERS = 3  # クラスタ数 (aruki, kudari, nobori の3種類)

# --- 1. データ読み込みと準備 ---
print(f"--- 1. データの読み込みと準備 ---")
try:
    df = pd.read_csv(INPUT_FILE)
except FileNotFoundError:
    print(f"エラー: 入力ファイル '{INPUT_FILE}' が見つかりません。")
    print("先に特徴量抽出のスクリプトを実行してください。")
    exit()

# 特徴量 (X) と 正解ラベル (y_true) に分割
# 'source_file' は分析に使用しないため除外
X = df.drop(['label', 'source_file'], axis=1)
y_true = df['label']

# ラベルのユニークな値を取得（後のプロットや評価で使用）
labels_true_unique = sorted(y_true.unique())

print(f"データセットの形状: {df.shape}")
print(f"特徴量の数: {X.shape[1]}")
print(f"正解ラベルの種類: {labels_true_unique}")

# ★★★ ここから追加 ★★★
print("\n--- 使用した特徴量一覧 ---")
# X.columnsで特徴量の名前のリストを取得し、番号付きで表示
for i, feature_name in enumerate(X.columns):
    print(f"{i+1:2d}: {feature_name}")
# ★★★ ここまで追加 ★★★

print("-" * 30, "\n")


# --- 2. 前処理 (標準化) ---
print(f"--- 2. 前処理 (標準化) ---")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("特徴量の標準化が完了しました。")
print("-" * 30, "\n")


# --- 3. K-meansクラスタリングの実行 ---
print(f"--- 3. K-meansクラスタリングの実行 (k={N_CLUSTERS}) ---")
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=42, n_init=10)
y_pred_cluster = kmeans.fit_predict(X_scaled) # 予測されたクラスタラベル (0, 1, 2)
df['cluster'] = y_pred_cluster
print("K-meansによるクラスタリングが完了しました。")
print("-" * 30, "\n")


# --- 4. 結果の評価 ---
print(f"--- 4. 結果の評価 ---")

# 4-1. K-meansクラスタと正解ラベルの対応付け
labels_pred_mapped = np.zeros_like(y_pred_cluster, dtype=object)
mapping = {}
for i in range(N_CLUSTERS):
    mask = (y_pred_cluster == i)
    if np.any(mask):
        most_common_label = y_true[mask].value_counts().idxmax()
        labels_pred_mapped[mask] = most_common_label
        mapping[i] = most_common_label
    else:
        mapping[i] = "Unassigned" 

print("クラスタと正解ラベルの対応付け:")
print(mapping)

# 4-2. 正答率 (Accuracy) の計算
accuracy = accuracy_score(y_true, labels_pred_mapped)
print(f"\n正答率 (Accuracy): {accuracy:.4f} ({accuracy*100:.2f}%)")

# 4-3. Adjusted Rand Index (ARI) の計算 (ラベル対応付け不要の評価指標)
ari = adjusted_rand_score(y_true, y_pred_cluster)
print(f"Adjusted Rand Index (ARI): {ari:.4f}")

# 4-4. 混同行列 (Confusion Matrix) の表示
conf_mat = confusion_matrix(y_true, labels_pred_mapped, labels=labels_true_unique)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', 
            xticklabels=labels_true_unique, yticklabels=labels_true_unique)
plt.title('Confusion Matrix (K-means)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
# plt.savefig('confusion_matrix_kmeans.png')
print("\n混同行列を 'confusion_matrix_kmeans.png' として保存しました。")
print("-" * 30, "\n")


# --- 5. 可視化 (PCAによる次元削減) ---
print(f"--- 5. 結果の可視化 ---")
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 可視化用のDataFrameを作成
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['True Label'] = y_true
df_pca['K-means Cluster'] = df['cluster'].map(mapping) 

plt.figure(figsize=(16, 7))

#  subplot 1: 元の正解ラベルによるプロット
plt.subplot(1, 2, 1)
sns.scatterplot(x='PC1', y='PC2', hue='True Label', hue_order=labels_true_unique, data=df_pca, palette='viridis', alpha=0.8, s=50)
plt.title('Original True Labels (after PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)

# subplot 2: K-meansの結果によるプロット
plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='K-means Cluster', hue_order=labels_true_unique, data=df_pca, palette='viridis', alpha=0.8, s=50)
plt.title('K-means Clustering Results (after PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)

plt.suptitle('K-means Clustering vs True Labels', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
# plt.savefig('kmeans_vs_true_labels_pca.png')
print("PCAによる可視化結果を 'kmeans_vs_true_labels_pca.png' として保存しました。")

plt.show()
```
